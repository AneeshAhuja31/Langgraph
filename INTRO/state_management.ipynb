# State Management
Langgraph introduces two key ideas to state management:
- State Schemas: Define the type of object given to each of the graph's node
- Reducers: They define how to apply Node outputs to the current state

A state is a dictionary 

## State
It is a shared data structure that holds the current information or context of the entire application. 

It is like the application's memory, keeping track of the variables and data that nodes can access as they execute.

## Nodes

They are individual functions or operations that perform specific tasks within the graph.

Each node receives input (often the current state), processes it and produces an output or an updated state.


## Graphs
It is the overarching structure that maps out how different nodes are connected and executed

It visually represents the workflow, showing the sequence and conditional paths between various operations

## Edges
They are the connection between nodes that determine the flow of execution

They tell us which node should be executed next after the current one completes its task.

## Condtional Edges
They decide the next node to execute based on specific conditions or logic applied to current state

## Start
It is a virtual entry point where the workflow begins.

It doesnt perform any operations itself but serves as the designated starting position for exectuion

## End 
Signifies the conclusion of the workflow in LangGraph

On reaching this node, the graph's execution stops.

## Tools
They are specialized functions or utilities that nodes can utilize to perform specific tasks such as fetching data from an api

They enhance capability of nodes by providing additional functionality.

Nodes are part of graph structure whereas tools are functionalities used within nodes

## Tool Node
A tool node is just a special kind of node whose main job is to run a tool

It connects the tool's output back into the State, so other nodes can use that information

## State Graph 
It is a class in LangGraph used to build and compile the graph structure

It manages nodes, edges and the overall state, ensuring that the workflow operates in a unified way and data flows correctly between components.

## Runnable
It is standardized, exectuable component that performs a specific task within an AI workflow

It serves as a fundamental building block, allowing for us to create modular systems

## Messages

**Human Message**: Represents input from a user.

**System Message**: Used to provide instructions or context to the model.

**Function Message**: Represents the result of a function call.

**AI Message**: Represents responses generated by AI models

**Tool Message**: Similar to function message but specific to tool usage
